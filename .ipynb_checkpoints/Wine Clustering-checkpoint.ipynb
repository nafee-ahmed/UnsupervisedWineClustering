Machines cannot interpret empty or null values. So, there are various methods to deal with it and they are:

* Removal of rows: The rows containing missing values are dropped. However, if there are lots of missing values in a dataset where there is not enough data already, then imputation is preferred.
* Imputation: Imputation involves filling in the missing values with estimated or imputed values. This can be done using various techniques, such as:
    - Mean/Median/Mode Imputation: Replace missing values with the mean, median, or mode of the non-missing values in the respective column.
    - Regression Imputation: Use regression models to predict missing values based on other features in the dataset.
    - K-Nearest Neighbors (KNN) Imputation: Fill in missing values based on the values of the nearest neighbors in the feature space.
    - Multiple Imputation: Generate multiple imputed datasets and use them for analysis, incorporating the uncertainty introduced by imputation.
* Indicator Variable: Create an additional binary column that indicates whether a value is missing or not. This allows the model to capture any patterns or relationships associated with missing values as a separate feature.

Usually removal of rows, and imputation solves the problem.

# Evaluating the 6 chosen clustering algorithms

The silhouette score is a metric to evaluate the quality of clustering results.Thus, the bigger the silhouette the score, the better the clustering. It ranges from -1 to 1. 

A silhouette score close to 1 indicates that the samples are well clustered, with clear separation between different clusters. A score of 0 indicates overlapping or ambiguous clusters, where samples may be assigned to multiple clusters. A negative score suggests that the samples are wrongly clustered, with samples assigned to clusters they do not belong to.

The 6 chosen clustering algorithms are:
* KMeans: is an iterative clustering algorithm that partitions the dataset into K distinct clusters. It aims to minimize the sum of squared distances between the data points and their cluster centroids.
* Agglomerative Clustering: is a hierarchical clustering algorithm that starts with each data point as a separate cluster and iteratively merges the closest clusters until a desired number of clusters is obtained. It builds a hierarchy of clusters using a linkage criterion such as Ward, average, or complete linkage.
* Spectral Clustering: is a graph-based clustering algorithm that uses the eigenvalues and eigenvectors of a similarity matrix derived from the data. It transforms the data into a lower-dimensional space and applies K-means or another clustering algorithm to cluster the transformed data.
* Birch: is a hierarchical clustering algorithm that constructs a tree-like structure to represent the data. It uses a combination of clustering and dimensionality reduction techniques to efficiently cluster large datasets
* Mean Shift: is a non-parametric clustering algorithm that does not require specifying the number of clusters in advance. It starts with an initial set of cluster centers and iteratively shifts them towards the modes of the underlying probability density function.
* Affinity Propagation: is a clustering algorithm that identifies exemplars in the data and assigns each data point to one of these exemplars. It uses a message-passing technique to exchange information between data points and determine the exemplars based on the similarity or affinity between data points.

```python
def train_and_score(models, df):
    np.random.seed(333) # makes the initial random number to be reproducible, so results are reproducible
    silhouettes = {} # stores the silhouette scores of different models
    for name, model in models.items(): # iterate over each model
        model.fit(df) # train each model 
        silhouettes[name] = silhouette_score(df, model.labels_) # calculate the silhouette score
    return silhouettes # return the silhouette scores
```

```python
# instantiate all the models and store them to be passed onto the train_and_score function
models = {'KMeans': KMeans(n_clusters=3, random_state=42), 
          'Agglo': AgglomerativeClustering(n_clusters=3),
          'Spectral': SpectralClustering(n_clusters=3, random_state=42),
          'Birch': Birch(n_clusters=3),
          'Mean Shift': MeanShift(),
          'Affinity': AffinityPropagation()}
# 'silhouettes' store the dictionary of silhouette scores of each model using train_and_score
silhouettes = train_and_score(models, df)
# convert the silhouettes to dataframe
model_compare = pd.DataFrame(silhouettes, index=["Silhouette Scores"])
ax = model_compare.T.plot.bar() # Plot a bar graph to plot the silhouette scores
plt.ylabel("Silhouette Scores")  # Add a label to the y-axis
# adding in the exact silhouette score on top of each bar
for p in ax.patches:
    ax.annotate(f"{p.get_height():.4f}", (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')
plt.show() # display the graph
```

# Choosing the best model

As the silhouette score is highest with the KMeans Algorithm, this is the model that performs the best with this dataset. So, it is chosen as the best model, in this case, and is then proceeded with. 

```python
kmeans = KMeans(n_clusters=3, random_state=42) # create k-means algorithm with 3 clusters
labels = kmeans.fit_predict(df) # train the model on the dataset and then send predictions
centroids = kmeans.cluster_centers_ # get coordinates of the centroid for clusters
centroids_pca = pca_2.transform(centroids) # transform the centroids using PCA
pd.Series(labels).value_counts() # count the number of data points in each cluster
```

```python
fig, ax = plt.subplots() # create instance of fig and ax for the plot
# add 2D reduced datapoints and color them based on clustering labels onto the scatter plot
ax.scatter(pca_2_res[:,0], pca_2_res[:,1], c=labels) 
# add centroids with marker 'x' with red color onto the scatter plot
ax.scatter(x=centroids_pca[:,0], y=centroids_pca[:,1], marker='x', color='red', s=169, linewidths=3)
# set the labels and title of the plot
ax.set(xlabel='X', ylabel='Y', title='Data represented by 2-dims')
# show the plot
plt.show()
```
