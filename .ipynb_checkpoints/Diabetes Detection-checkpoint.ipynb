```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
print(df.info())
print(f'The shape of the dataset: {df.shape}')
```

```python
df.nunique() # number of unique values in each column
```

```python
# identifies ranges of values to detect outliers
def plot_box_plot(df):
    fig, ax = plt.subplots(1, figsize=(17,4))
    ax = sns.boxplot(data=df)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    plt.show()
plot_box_plot(df)
print(f"Number of outliers in the column Insulin: {df[df['Insulin'] > 600]['Insulin'].count()}")
```

In the cell above, we can see the column, insulin, has 3 outliers, so let's remove them

```python
df = df[df['Insulin'] <= 600]
# after removal of outliers
plot_box_plot(df)
```

There are different ranges of values across all the columns. So let's normalize. So z-score is done on all the columns except the target column

```python
cols_zscore = df.keys()
cols_zscore = cols_zscore.drop('Outcome')
df[cols_zscore] = df[cols_zscore].apply(stats.zscore)
# after normlization
plot_box_plot(df)
```
